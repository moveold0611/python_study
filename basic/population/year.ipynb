```python
import pandas as pd
import matplotlib as plt
```

```python
data = pd.read_csv("population_data.csv", encoding="cp949")
```

```python
dataDict = {}
for i in range(1, len(data)):
    dataDict[data.iloc[i, 0]] = data.iloc[1, range(i, len(data.columns), 6)]

print(dataDict)
```

```python
name = input("행정구역을 입력하세요: ")
for i, j in  dataDict.items(): # key, values를 출력
    if name in i:
        print(i)
        print(dataDict.get(i))

```
