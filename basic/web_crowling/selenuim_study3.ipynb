```python
# ê²€ìƒ‰ì–´ [ì•„ì´í°15, ì†í¥ë¯¼, react, ì•¼êµ¬]
# list > dict > list > dict  /   searchValue, data_list[{title / hits / creator / description / url}]

from selenium import webdriver # í¬ë¡¬ ë“œë¼ì´ë²„
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import numpy as np
import time

url = "https://www.youtube.com/"
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)

data = {}
dataList = [] 
contentList = []
driver = webdriver.Chrome(options=chrome_options)
driver.get(url)
time.sleep(5)

searchValue = ["ì•„ì´í°15", "ì†í¥ë¯¼", "react", "ì•¼êµ¬"]
searchInput = driver.find_element(by=By.CSS_SELECTOR, value='input[name="search_query"]')

time.sleep(2)
for i in range(len(searchValue)):
    searchInput.send_keys(searchValue[i])
    searchInput.send_keys(Keys.RETURN)
    time.sleep(5)

    contentbox = driver.find_element(by=By.ID, value="contents")
    contents = driver.find_elements(by=By.CSS_SELECTOR, value='ytd-video-renderer[class="style-scope ytd-item-section-renderer"]')
    
    j = 0
    for content in contents:
        title = content.find_element(by=By.CSS_SELECTOR, value='yt-formatted-string[class="style-scope ytd-video-renderer"]').text
        hits = content.find_element(by=By.CSS_SELECTOR, value='span[class="inline-metadata-item style-scope ytd-video-meta-block"]').text
        creator = content.find_element(by=By.CSS_SELECTOR, value='div[id="channel-info"] ytd-channel-name[id="channel-name"] a').text
        description = content.find_element(by=By.CSS_SELECTOR, value='div div yt-formatted-string').text
        url = content.find_element(by=By.CSS_SELECTOR, value='a[id="video-title"]').get_attribute(name="href")

        if hits[:1] == "ì¡°":
            hits = hits.split(" ")[1].split("íšŒ")[0]
            if hits[-1:] == "ë§Œ":
                hits = hits.split("ë§Œ")[0]
                hits = float(hits) * 10000
                hits = str(hits)[0:-2]
            elif hits[-1:] == "ì²œ":
                hits = hits.split("ì²œ")[0]
                hits = float(hits) * 1000
                hits = str(hits)
        else:
            hits = "ì‹¤ì‹œê°„"

        contentDict = {
            'title': title,
            'hits': hits,
            'creator': creator,
            'description': description,
            'url': url
        }

        contentList.append(contentDict)

        data = {
            'searchValue': searchValue[i],
            'dataList': contentList
        }
        contentList = []
        dataList.append(data)
 
        j += 1
        if j == 10:
            break

    searchInput.clear()
    time.sleep(0.5)

# dataNp = np.array(dataList)

dataPd = pd.DataFrame(dataList)
print(dataPd)
```
