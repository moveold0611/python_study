```python

import pandas as pd
import numpy as np

weatherDf = pd.read_csv("weather_data.csv", encoding='cp949')

# 전처리 (필요없는 항목 필터링)
weatherDf = weatherDf[['지점명', "일시", '기온(°C)']]
weatherDf
```

```python
#전처리2-1 (필요한 카테고리 선택)
is_seoul = weatherDf["지점명"] == "서울"
seoul = weatherDf[is_seoul]
seoul
```

```python
#전처리2-2
is_busan = weatherDf["지점명"] == "부산"
busan = weatherDf[is_busan]
busan
```

```python
#전처리3  (누락된 항목 갯수 일치시키기)

# 서울
temp = []
for date in seoul["일시"]:
    temp.append(date.split(" ")[0]) # '날짜 시간'형태의 자료를 " "공백으로 잘라서 날짜만 [0]인덱스로 append
temp_seoul = []
unique_temp = set(temp) #set 자료형으로 형변환하여 중복자료 제거


for date in unique_temp:
    if temp.count(date) != 24:
        temp_seoul.append(date) # 누락된 데이터 찾기 (8월 26일)

print(temp_seoul)
timeNp = np.arange(24)
findTime = []
full_date = []
result_seoul = []
array_seoul =[]
# 8월 26일 중 누락된 시간 찾기
for date in seoul["일시"]: 
    for temp_date in temp_seoul:
        if(temp_date in date):
            findTime.append(date.split(" ")[1])

print(findTime)

# 누락된 시간만 정제
for i in range(len(temp_seoul)):
    print(temp_seoul[i])
    array_seoul.clear()
    for j in timeNp:
        full_date.append(temp_seoul[i] + " " + str(j).zfill(2)+ ":00")

    for j in range(len(findTime)):
        array_seoul.append(temp_seoul[i] + " " + findTime[j])

    for j in timeNp:
        if full_date[j] != array_seoul[j]:
            result_seoul.append(full_date[j])
            array_seoul.insert(j, full_date[j])

print(result_seoul)
```

```python
# 부산
temp = []
for date in busan["일시"]:
    temp.append(date.split(" ")[0])
temp_busan = []
unique_temp = set(temp)


for date in unique_temp:
    if temp.count(date) != 24:
        temp_busan.append(date)

timeNp = np.arange(24)
findTime = []
full_date = []
result_busan = []
array_busan =[]

for date in busan["일시"]: 
    for temp_date in temp_busan:
        if(temp_date in date):
            findTime.append(date.split(" ")[1])
            

# 누락된 시간만 정제
for i in range(len(temp_busan)):
    array_busan.clear()
    for j in timeNp:
        full_date.append(temp_busan[i] + " " + str(j).zfill(2)+ ":00")

    for j in range(len(findTime)):
        array_busan.append(temp_busan[i] + " " + findTime[j])

    for j in timeNp:
        if full_date[j] != array_busan[j]:
            result_busan.append(full_date[j])
            array_busan.insert(j, full_date[j])

print(result_busan)
```

```python
full_date = list(set(result_seoul + result_busan))

for date in full_date:
    seoul = seoul[seoul["일시"] != date]
    busan = busan[busan["일시"] != date]
print(seoul)
time = []
for i in range(24):
    time.append(str(i).zfill(2) + ":00")

zeros = np.zeros((24, 2))
df = pd.DataFrame(zeros, index=time, columns=["seoul", 'busan'])

df
```
